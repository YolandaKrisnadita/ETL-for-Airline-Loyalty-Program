# Connect to Datasource

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-airlane-loyalty-program'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'airlane-loyalty-program'
path_to_data = '/Users/yoland-mac/Hacktiv8Jan/Phase 2/Milestone 3/p2-ftds003-sby-m3-YolandaKrisnadita/project_m3/dags/P2M3_Yolanda-Krisnadita_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

# Create Expectation Suite

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-airlane-loyalty-program'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

# Expectations

## Expectation 1 to be unique

```python
# Expectation 1 : Column `id` must be unique as it's a history primary key:
validator.expect_column_values_to_be_unique('id')
```

## Expectation 2 to be between min_value and max_value

```python
# Expectation 2: Column `enrollment_year` must in between of the specifict year
validator.expect_column_values_to_be_between(
    column='enrollment_year', min_value=2012, max_value=2018
)
```

## Expectation 3 to be in set

```python
# Expectation 3 : Column `loyalty_card` must contain one of the following 3 things :
# Star, Nova and Aurora

validator.expect_column_values_to_be_in_set('loyalty_card', ['Star','Nova','Aurora'])
```

## Expectation 4 to be in type list

```python
# Expectation 4 : Column `points_accumulated` must be in type of list.
validator.expect_column_values_to_be_in_type_list('points_accumulated', ['integer', 'float'])
```

## Expectation 5 to be valid country

```python
#Expectation 5 : Column `country` must be in Canada (6 Lenght)

validator.expect_column_value_lengths_to_equal('country',6)
```

## Expectation 6 to be not null

```python
#Expectation 6: Column `cancellation_year` must not be null
validator.expect_column_values_to_not_be_null('cancellation_year')
```

## Expectation 7 to be in Quartile to be between posistive Skew

```python
#Expectation 7 : Column `clv` must in quartile to be between positif skew
validator.expect_column_quantile_values_to_be_between('clv',{'quantiles': [0.25, 0.75],'value_ranges': [ [-3545.42598045427, 5682.606151909146], [5682.606151909146, 14910.638284272562]]})
```

# Save Expectation Suite

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```
